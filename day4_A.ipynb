**seperate the input and output variables**

**import the class**

```python
from sklearn.linear_model import LinearRegression
```

**create the object**

**DECISION TREE CLASSIFICATION**

**import the dataset**

```python
df = pd.read_csv('Social_Network_Ads.csv')
```

```python
df.shape
```

```python
list(df.columns)
```

```python
df
```

**seperate input and output variables**

```python
x = df[['Age','EstimatedSalary']]
y = df['Purchased']
```

**explore the data**

```python
plt.title('Purchase Data')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.grid()
plt.scatter(x['Age'], x['EstimatedSalary'],c=y)
```

**import the class**

```python
from sklearn.tree import DecisionTreeClassifier
```

**create object**

```python
Classifier = DecisionTreeClassifier(random_state=0)
```

**train the model**

```python
Classifier.fit(x,y)
```

**plot the tree**

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(12,15))
plot_tree(Classifier, fontsize=7, feature_names=['age','sal'],
          class_names=['y','n'], filled=True, rounded=True);
```

**Predict on new data**

```python
new=pd.DataFrame({
    'Age': [23,45,54,67],
    'EstimatedSalary':[34000,41000,26000,134000]
})
```

```python
new
```

```python
Classifier.predict(new)
```
